# Analysis of curve fitting data from CAP subset n = 30

Uses first 30 samples

```python
import numpy as np
import pandas

```

```python
cap_subset = pandas.read_csv('Z:/User Folders/SRajurkar/X3726/CAP Baseline subset/final csv.csv')

cap_subset.head()
```

Correlation between MZ curve and 15.14

```python
cap_subset.plot.scatter(x='MZ curve', y='15.14')
x = cap_subset['MZ curve']
y = cap_subset['15.14']
r = np.corrcoef(x, y)

print('r = ',  r[0][1])
```

Correlation between MZ curve and MZ dropdown bin

```python
cap_subset.plot.scatter(x='MZ curve', y='MZ bin')

x = cap_subset['MZ curve']
y = cap_subset['MZ bin']
r = np.corrcoef(x, y)

print('r = ', r[0][1])
```
